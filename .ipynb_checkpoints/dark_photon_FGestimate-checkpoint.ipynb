```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
fg_857 = hp.read_map("Darkphoton_FG_estimate/HFI_CompMap_Foregrounds-commander-857_R3.00.fits")
full_857 = hp.read_map("Darkphoton_FG_estimate/HFI_SkyMap_857-field-Int_2048_R3.00_full.fits")
fg_545 = hp.read_map("Darkphoton_FG_estimate/HFI_CompMap_Foregrounds-commander-545_R3.00.fits")
full_545 = hp.read_map("Darkphoton_FG_estimate/HFI_SkyMap_545-field-Int_2048_R3.00_full.fits")
fg_353 = hp.read_map("Darkphoton_FG_estimate/HFI_CompMap_Foregrounds-commander-353_R3.00.fits")
full_353 = hp.read_map("Darkphoton_FG_estimate/HFI_SkyMap_353-psb-field-IQU_2048_R3.00_full.fits")
fg_217 = hp.read_map("Darkphoton_FG_estimate/HFI_CompMap_Foregrounds-commander-217_R3.00.fits")
full_217 = hp.read_map("Darkphoton_FG_estimate/HFI_SkyMap_217-field-IQU_2048_R3.00_full.fits")
fg_143 = hp.read_map("Darkphoton_FG_estimate/HFI_CompMap_Foregrounds-commander-143_R3.00.fits")
full_143 = hp.read_map("Darkphoton_FG_estimate/HFI_SkyMap_143-field-IQU_2048_R3.00_full.fits")
fg_100 = hp.read_map("Darkphoton_FG_estimate/HFI_CompMap_Foregrounds-commander-100_R3.00.fits")
full_100 = hp.read_map("Darkphoton_FG_estimate/HFI_SkyMap_100-field-IQU_2048_R3.00_full.fits")
full_30 = hp.pixelfunc.ud_grade(hp.read_map("Darkphoton_FG_estimate/LFI_SkyMap_030-BPassCorrected-field-IQU_1024_R3.00_full.fits"),2048)
fg_30 = hp.pixelfunc.ud_grade(hp.read_map("Darkphoton_FG_estimate/LFI_CompMap_Foregrounds-commander-030_R3.00.fits"),2048)
full_44 = hp.pixelfunc.ud_grade(hp.read_map("Darkphoton_FG_estimate/LFI_SkyMap_044-BPassCorrected-field-IQU_1024_R3.00_full.fits"),2048)
fg_44 = hp.pixelfunc.ud_grade(hp.read_map("Darkphoton_FG_estimate/LFI_CompMap_Foregrounds-commander-044_R3.00.fits"),2048)
full_70 = hp.pixelfunc.ud_grade(hp.read_map("Darkphoton_FG_estimate/LFI_SkyMap_070-BPassCorrected-field-IQU_1024_R3.00_full.fits"),2048)
fg_70 = hp.pixelfunc.ud_grade(hp.read_map("Darkphoton_FG_estimate/LFI_CompMap_Foregrounds-commander-070_R3.00.fits"),2048)

cmb = hp.read_map("COM_CMB_IQU-smica_2048_R3.00_full.fits")

mask = hp.read_map("HFI_Mask_GalPlane-apo0_2048_R2.00.fits",field=1)
mask_hi = hp.read_map("HFI_Mask_GalPlane-apo2_2048_R2.00.fits",field=1)

masked_30 = hp.ma(full_30)
masked_30.mask = np.logical_not(mask)
masked_44 = hp.ma(full_44)
masked_44.mask = np.logical_not(mask)
masked_70 = hp.ma(full_70)
masked_70.mask = np.logical_not(mask)
masked_100 = hp.ma(full_100)
masked_100.mask = np.logical_not(mask)
masked_143 = hp.ma(full_143)
masked_143.mask = np.logical_not(mask)

masked_cmb = hp.ma(cmb)
masked_cmb.mask = np.logical_not(mask)

f_857 = 1./np.sqrt(5.19)
f_545 = 1./np.sqrt(3365)

Bl30 = hp.sphtfunc.gauss_beam(0.0094247, lmax=3*2048-1, pol=False)
Bl44 = hp.sphtfunc.gauss_beam(0.0078830, lmax=3*2048-1, pol=False)
Bl70 = hp.sphtfunc.gauss_beam(0.00387463, lmax=3*2048-1, pol=False)
Bl100 = hp.sphtfunc.gauss_beam(0.0028187, lmax=3*2048-1, pol=False)
Bl143 = hp.sphtfunc.gauss_beam(0.0021234, lmax=3*2048-1, pol=False)
Bl217 = hp.sphtfunc.gauss_beam(0.0014602, lmax=3*2048-1, pol=False)
Bl353 = hp.sphtfunc.gauss_beam(0.0014369, lmax=3*2048-1, pol=False)
Bl545 = hp.sphtfunc.gauss_beam(0.0014049, lmax=3*2048-1, pol=False)
Bl857 = hp.sphtfunc.gauss_beam(0.0013497, lmax=3*2048-1, pol=False)
```

```python
hp.mollview((full_30-fg_30)*mask)
hp.mollview(cmb*mask,min=-.0005,max=.0005)
```

```python
pspec=hp.anafast((full_30)*mask)
pspec_cmb = hp.anafast(mask*cmb)
```

```python
plt.loglog(pspec/pspec_cmb)
pspec_cmb[20]/pspec[20]
```

```python
pspec_30 = hp.anafast(masked_30)
pspec_44 = hp.anafast(masked_44)
pspec_70 = hp.anafast(masked_70)
pspec_100 = hp.anafast(masked_100)
pspec_143 = hp.anafast(masked_143)
pspec_217 = hp.anafast(mask_hi*full_217)
pspec_353 = hp.anafast(mask_hi*full_353)
pspec_545 = hp.anafast(mask_hi*full_545*f_545)
pspec_857 = hp.anafast(mask_hi*full_857*f_857)
pspec_cmb = hp.anafast(mask*cmb)
```

```python
tcmb=2.726e6
fsky=.4
plt.loglog(tcmb**2*pspec_30/Bl30/Bl30/fsky)
plt.loglog(tcmb**2*pspec_44/Bl44/Bl44/fsky)
plt.loglog(tcmb**2*pspec_70/Bl70/Bl70/fsky)
plt.loglog(tcmb**2*pspec_100/Bl100/Bl100/fsky)
plt.loglog(tcmb**2*pspec_143/Bl143/Bl143/fsky)
plt.loglog(tcmb**2*pspec_217/Bl217/Bl217/fsky)
plt.loglog(tcmb**2*pspec_353/Bl353/Bl353/fsky)
plt.loglog(tcmb**2*pspec_545/Bl545/Bl545/fsky)
plt.loglog(tcmb**2*pspec_857/Bl857/Bl857/fsky)
plt.loglog(tcmb**2*pspec_cmb/fsky)
plt.ylim(1e-3,1e13)
```

```python
Planck_pspec = np.zeros((9,pspec_30.shape[0]))
Planck_pspec[0,:] = tcmb**2*pspec_30/Bl30/Bl30
Planck_pspec[1,:] = tcmb**2*pspec_44/Bl44/Bl44
Planck_pspec[2,:] = tcmb**2*pspec_70/Bl70/Bl70
Planck_pspec[3,:] = tcmb**2*pspec_100/Bl100/Bl100
Planck_pspec[4,:] = tcmb**2*pspec_143/Bl143/Bl143
Planck_pspec[5,:] = tcmb**2*pspec_217/Bl217/Bl217
Planck_pspec[6,:] = tcmb**2*pspec_353/Bl353/Bl353
Planck_pspec[7,:] = tcmb**2*pspec_545/Bl545/Bl545
Planck_pspec[8,:] = tcmb**2*pspec_857/Bl857
np.save('Planck_spectra',Planck_pspec)
```

```python
coefficients = tcmb*tcmb*np.array([2.5e-9,1.5e-8,5e-7,2e-4,9e-1])/fsky
frequencies = np.array([143,217,353,545,857])
new_frequencies = np.array([145,225,278])
fit_interp = np.interp(new_frequencies,frequencies,coefficients)

ls=np.arange(pspec_cmb.shape[0])
plt.loglog(ls,fit_interp[0]*ls**-3.)
plt.loglog(ls,fit_interp[1]*ls**-3.)
plt.loglog(ls,fit_interp[2]*ls**-3.)
plt.loglog(tcmb*tcmb*pspec_cmb/fsky)
```

```python
# Planck frequencies

cib_100 = hp.read_map("websky_CIB/cib_nu0100.fits",nest=True)*(4.1877e3)
cib_100_dg = hp.pixelfunc.ud_grade(cib_100,2048)
cib_143 = hp.read_map("websky_CIB/cib_nu0143.fits",nest=True)*(2.6320e3)
cib_143_dg = hp.pixelfunc.ud_grade(cib_143,2048)
cib_217 = hp.read_map("websky_CIB/cib_nu0217.fits")*(2.0676e3)
cib_217_dg = hp.pixelfunc.ud_grade(cib_217,2048)
cib_353 = hp.read_map("websky_CIB/cib_nu0353.fits",nest=True)*(3.371e3)
cib_353_dg = hp.pixelfunc.ud_grade(cib_353,2048)
cib_545 = hp.read_map("websky_CIB/cib_nu0545.fits",nest=True)*(1.7508e4)
cib_545_dg = hp.pixelfunc.ud_grade(cib_545,2048)
cib_857 = hp.read_map("websky_CIB/cib_nu0857.fits",nest=True)*(6.9653e5)
cib_857_dg = hp.pixelfunc.ud_grade(cib_857,2048)
```

```python
pcib_100 = hp.anafast(cib_100_dg)
pcib_143 = hp.anafast(cib_143_dg)
pcib_217 = hp.anafast(cib_217_dg)
pcib_353 = hp.anafast(cib_353_dg)
pcib_545 = hp.anafast(cib_545_dg)
pcib_857 = hp.anafast(cib_857_dg)
```

```python
# S4 Frequencies

cib_93 = hp.read_map("websky_CIB/cib_nu0093.fits",nest=True)*(4.6831e3)
cib_145 = hp.read_map("websky_CIB/cib_nu0145.fits",nest=True)*(2.5947e3)
cib_225 = hp.read_map("websky_CIB/cib_nu0225.fits",nest=True)*(2.0716e3)
cib_278 = hp.read_map("websky_CIB/cib_nu0278.fits",nest=True)*(2.3302e3)

```

```python
pcib_93 = hp.anafast(cib_93)
pcib_145 = hp.anafast(cib_145)
pcib_225 = hp.anafast(cib_225)
pcib_278 = hp.anafast(cib_278)
```

```python
Bl93 = hp.sphtfunc.gauss_beam(0.000727221, lmax=3*4096-1, pol=False)
Bl145 = hp.sphtfunc.gauss_beam(0.000465421, lmax=3*4096-1, pol=False)
Bl225 = hp.sphtfunc.gauss_beam(0.000319977, lmax=3*4096-1, pol=False)
Bl278 = hp.sphtfunc.gauss_beam(0.000290888, lmax=3*4096-1, pol=False)

ls=np.arange(pcib_93.shape[0])

plt.loglog((pcib_93)/Bl93/Bl93)
plt.loglog((pcib_145+ fit_interp[0]*ls**-3.)/Bl145/Bl145)
plt.loglog((pcib_225 + fit_interp[1]*ls**-3.)/Bl225/Bl225)
plt.loglog((pcib_278 + fit_interp[2]*ls**-3.)/Bl278/Bl278)

plt.loglog(pspec_cmb*tcmb*tcmb/.4)
```

```python
S4_pspec = np.zeros((4,pcib_93.shape[0]))
S4_pspec[0,:] = (pcib_93)/Bl93/Bl93
S4_pspec[1,:] = (pcib_145+ fit_interp[0]*ls**-3.)/Bl145/Bl145
S4_pspec[2,:] = (pcib_225 + fit_interp[1]*ls**-3.)/Bl225/Bl225
S4_pspec[3,:] = (pcib_278 + fit_interp[2]*ls**-3.)/Bl278/Bl278
np.save('S4_spectra',S4_pspec)
```
